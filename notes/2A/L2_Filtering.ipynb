```python
# load an image
img = cv2.imread(img_root_path + 'bicycle.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# Applying Gaussian filter
h_size = 31
sigma = 5
blur_img = cv2.GaussianBlur(img, (h_size,h_size), sigma)
```

```python
plt.imshow(blur_img, cmap='gray')
plt.show()
```

```python
# larger sigma
sigma2 = 10
blur_img2 = cv2.GaussianBlur(img, (h_size, h_size), sigma2)
plt.imshow(blur_img2, cmap='gray')
plt.show()
```

# Removing noise using filter

```python
saturn_img = cv2.imread(img_root_path + 'saturn.png', cv2.IMREAD_GRAYSCALE)
img_height = saturn_img.shape[0]
img_width = saturn_img.shape[1]
print(img_width, img_height)
plt.imshow(saturn_img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# add some noise to the image
noise_sigma = 25

np.random.seed(999)
mean, std = 0, 1
noise = np.random.normal(mean, std, (img_height,img_width)) * noise_sigma
noisy_saturn_img = saturn_img + noise
plt.imshow(noisy_saturn_img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# Create a Guassian filter and remove noise
filter_size = (11,11)
filter_sigma = 2

filtered_saturn = cv2.GaussianBlur(noisy_saturn_img, filter_size,filter_sigma)
plt.imshow(filtered_saturn, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Therefore, when filtering with a Gaussian, these are true
1. The sigma is most imporant - it defines the blur kernel's scale with respect to the image.  
2. Altering the normalization coefficient does not effect the blur, only the brightness
