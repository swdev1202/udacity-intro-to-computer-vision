# Edges
- Finding edges => finding maxima/minima in derivatives

## Differential Operators
* differential operators - when applied to the image returns some derivatives.
* modle these "operators" as masks/kernels that compute the image graident function.
* threshold this gradient function to select the edge pixels

## Graident
* the gradient points in the direction of most rapid increase in intensity

## The discrete gradient
* we want an "operator" (mask/kernel) that we can apply to the image that implements   
$ \frac{\partial f(x,y)}{\partial x} = lim_{\varepsilon -> 0} \frac{f(x+\varepsilon, y) - f(x,y)}{\varepsilon}$
